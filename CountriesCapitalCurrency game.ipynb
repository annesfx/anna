```python

# Global variables for tracking score and hints
score = 0
hints_given = 0

def check_guess():
    global score
    selected_capital = capital_var.get()
    selected_currency = currency_var.get()
    country = current_country

    correct_capital = countries_info[country]["capital"]
    correct_currency = countries_info[country]["currency"]
    
    if selected_capital == correct_capital and selected_currency == correct_currency:
        messagebox.showinfo("Congratulations!", "You guessed it right!")
        score += 1
    else:
        messagebox.showerror("Incorrect!", f"The correct answer is Capital: {correct_capital}, Currency: {correct_currency}.")
    
    update_score_label()
    next_country()

def next_country():
    global hints_given, current_country
    current_country = random.choice(list(countries_info.keys()))
    country_label.config(text=f"Country: {current_country}")
    
    # Generate multiple choices for capitals and currencies
    capital_choices = [countries_info[current_country]["capital"]]
    currency_choices = [countries_info[current_country]["currency"]]
    
    while len(capital_choices) < 4:
        random_capital = random.choice(list(countries_info.values()))["capital"]
        if random_capital not in capital_choices:
            capital_choices.append(random_capital)
    
    while len(currency_choices) < 4:
        random_currency = random.choice(list(countries_info.values()))["currency"]
        if random_currency not in currency_choices:
            currency_choices.append(random_currency)
    
    random.shuffle(capital_choices)
    random.shuffle(currency_choices)
    
    for i in range(4):
        capital_options[i].config(text=capital_choices[i], value=capital_choices[i])
        currency_options[i].config(text=currency_choices[i], value=currency_choices[i])
    
    hints_given = 0  # Reset hints given for the new country

def give_hint():
    global hints_given
    country = current_country
    capital_hint = countries_info[country]["capital"]
    currency_hint = countries_info[country]["currency"]
    messagebox.showinfo("Hint", f"Capital: {capital_hint}, Currency: {currency_hint}")
    hints_given += 1

def update_score_label():
    score_label.config(text=f"Score: {score}")

# Create the main window
root = tk.Tk()
root.title("Country Capital and Currency Guessing Game")

# Label to display the country
country_label = tk.Label(root, text="", font=("Helvetica", 24), padx=10, pady=10)
country_label.pack()

# Variables to hold the selected values
capital_var = tk.StringVar()
currency_var = tk.StringVar()

# Capital multiple-choice options
capital_options = []
for _ in range(4):
    rb = tk.Radiobutton(root, text="", variable=capital_var, font=("Helvetica", 14))
    rb.pack(anchor='w')
    capital_options.append(rb)

# Currency multiple-choice options
currency_options = []
for _ in range(4):
    rb = tk.Radiobutton(root, text="", variable=currency_var, font=("Helvetica", 14))
    rb.pack(anchor='w')
    currency_options.append(rb)

# Button to check the guess
check_button = tk.Button(root, text="Check Guess", command=check_guess, font=("Helvetica", 14), bg="#D7BDE2")
check_button.pack()

# Button to give hint
hint_button = tk.Button(root, text="Know the history", command=give_hint, font=("Helvetica", 14), bg="#D7BDE2")
hint_button.pack()

# Label to display the score
score_label = tk.Label(root, text="Score: 0", font=("Helvetica", 14), bg="#D7BDE2")
score_label.pack()

# Button to move to the next country
next_button = tk.Button(root, text="Next Country", command=next_country, font=("Helvetica", 14), bg="#D7BDE2")
next_button.pack()

# Start the game with the first country
next_country()

# Run the Tkinter event loop
root.mainloop()
```
